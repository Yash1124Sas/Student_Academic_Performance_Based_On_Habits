```python
X=df.drop('exam_score',axis=True)
y=df['exam_score']
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
columns=['gender','part_time_job','diet_quality','parental_education_level','internet_quality','extracurricular_participation']
Xlabeled=X.copy()
label_encoders={}
for i in columns:
    Xlabeled=pd.get_dummies(Xlabeled,drop_first=True)
```

```python
Xlabeled.astype(int)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
Xlabeled=scaler.fit_transform(Xlabeled)
```

```python
Xlabeled
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import  r2_score,mean_absolute_percentage_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(Xlabeled, y, test_size=0.20, random_state=42)
```

```python
models={
    'linear Regression':LinearRegression(),
    'KNN':KNeighborsRegressor(),
    'Decision Tree':DecisionTreeRegressor(),
    'Random Forest':RandomForestRegressor(),
    'XGBOOST Regressor':XGBRegressor()
}
```

```python
result=[]
```

```python
for name, model in models.items():
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    r2=r2_score(y_test,y_pred)
    mape=mean_absolute_percentage_error(y_test,y_pred)
    y_pred_train=model.predict(X_train)
    r2_tr=r2_score(y_train,y_pred_train)
    mape_tr=mean_absolute_percentage_error(y_train,y_pred_train)
    result.append({
        "model":name,
        "R2 Score_Test":round(r2,4),
        "MAPE_Test":round(mape,4),
        "R2 Score_Train":round(r2_tr,4),
        "MAPE_Train":round(mape_tr,4)
    })
```

```python
print(result)
```

```python
import joblib
joblib.dump(models["linear Regression"],"linearRegression.pkl")
joblib.dump(scaler,"ScalerLR.pkl")
joblib.dump(X.columns.tolist(),'ColumnsLR.pkl')
```
